```python
#defining a function to view some more insights of the dataset.
def printinfo():
    temp = pd.DataFrame(index=data.columns)
    temp['data_type'] = data.dtypes
    temp['null_count'] = data.isnull().sum()
    temp['unique_count'] = data.nunique()
    return temp
printinfo()
```

From the above command it is noticed that row 10472 have a missing data for category column and all the prevailing column values is being replaced with its previous column. Better idea will be to drop this row from out dataframe.

```python
data.dropna(subset = ['Content Rating'], inplace=True)
```

Also dropping unwanted columns from the dataset:

```python
data.drop(['Current Ver','Last Updated', 'Android Ver'], axis=1, inplace=True)
data.head()
```

Replacing the missing values of the Rating Column with the Mode value of that entire column.

```python
modeValueRating = data['Rating'].mode()
modeValueRating[0]
```

```python
data['Rating'].fillna(value=modeValueRating[0], inplace = True)
printinfo()
```

Thus,completed the data cleaning process.

Before continuing with data preparation the columns 'Reviews','Size','Installs',&'Price' should be converted from object type  to integer or float datatype.

```python
#Converting the Reviews column to integer.
data['Reviews'] = data.Reviews.astype(int)
printinfo()
```

```python
#Converting the Size Column from object to integer:
```

```python
data['Size'] = data.Size.apply(lambda x: x.strip('+'))#removing the '+'symbol
```

```python
data['Size'] = data.Size.apply(lambda x: x.replace(',', ''))#removing the `,`symbol
```

```python
data['Size'] = data.Size.apply(lambda x: x.replace('M', 'e+6'))# For converting the M to Mega 
#Replacing the M by multiplying the value with 1000000
```

```python
data['Size'] = data.Size.apply(lambda x: x.replace('k', 'e+3'))# For convertinf the K to Kilo
#Replacing the k by multiplying the value with 1000
```

```python
data['Size'] = data.Size.replace('Varies with device', np.NaN)
#Replacing the Varies with device value with Nan.
```

```python
data['Size'] = pd.to_numeric(data['Size']) # Converting the string to Numeric type
printinfo()
```

```python
#dropping Rows of the column Size having Nan values
data.dropna(subset = ['Size'], inplace=True)
printinfo()
```

```python
#Converting the Installs column from object to integer
```

```python
data['Installs'] = data.Installs.apply(lambda x: x.strip('+')) #removing the '+'symbol
```

```python
data['Installs'] = data.Installs.apply(lambda x: x.replace(',', ''))#removing the `,`symbol
```

```python
data['Installs'] = pd.to_numeric(data['Installs'])#converting it from string type to numeric type
printinfo()
```

```python
#Converting the Price column from object to numeric
```

```python
data['Price'].value_counts()
```

```python
data['Price'] = data.Price.apply(lambda x: x.strip('$')) #removing the '$'symbol
```

```python
data['Price'] = pd.to_numeric(data['Price'])#converting the values to Numeric type.
printinfo()
```

```python
data.describe()#displays some basic statistical details like percentile, mean, std etc. of a data frame
# or a series of numeric values.
#Ignores the null values and object data types by default.
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
printinfo()
```

```python
data['Category'].value_counts()
```

#### Visualizing the plot:

#### •Top Categories

```python
y = data['Category'].value_counts().index
x = data['Category'].value_counts()
xasis = []
yasis = []
for i in range(len(x)):
    xasis.append(x[i])
    yasis.append(y[i])
```

```python
plt.figure(figsize=(25,15))
plt.xlabel("Count")
plt.ylabel("Category")

graph = sns.barplot(x = xasis, y = yasis, palette= "husl")
graph.set_title("Top categories in the Android Market", fontsize = 25);
```

##### From the above output we can conclude that in playstore most of the apps are under Family & Game category and least are of Beauty & Comics Category.

#### • Distribution of rating:

```python
data['Rating'].describe()
```

```python
plt.figure(figsize=(15,9))
plt.xlabel("Rating")
plt.ylabel("Frequency")
graph = sns.kdeplot(data.Rating, color="Blue", shade = True)
plt.title('Distribution of Rating',size = 20);
```

##### From the above graph we can come to a conclusion that most of the apps in google playstore are rated in between 3.5 to 4.8

#### • Highest installations

```python
Highest_Installs_data = data.groupby('Category')[['Installs']].sum().sort_values(by='Installs', ascending=False)
```

```python
Highest_Installs_data.head(10)
```

```python
x_asis = []
y_asis = []

for i in range(len(Highest_Installs_data)):
    x_asis.append(Highest_Installs_data.Installs[i])
    y_asis.append(Highest_Installs_data.index[i])

plt.figure(figsize=(18,13))

plt.xlabel("Installs")
plt.ylabel("Category")
graph = sns.barplot(x = x_asis, y = y_asis, alpha =0.9, palette= "viridis")
graph.set_title("Installs", fontsize = 25);
```

##### From the above visualization, it can be interpreted that the top categories with highest installs are Game, Family, Communication, News & Magazines, & Tools.

#### • Percent of Free Vs Paid Apps:

```python
plt.figure(figsize=(10,10))
labels = data['Type'].value_counts(sort = True).index
sizes = data['Type'].value_counts(sort = True)
colors = ["blue","cyan"]
explode = (0.2,0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=0)
plt.title('Percent of Free Vs Paid Apps in Playstore',size = 20)
plt.show()
```

#####  From the above graph we can see that 92%(Approx.) of apps in google play store are free and 8%(Approx.) are paid.

#### •Top Genres:

```python
topAppsinGenres = data['Genres'].value_counts().head(50)
```

```python
x = []
y = []

for i in range(len(topAppsinGenres)):
    x.append(topAppsinGenres.index[i])
    y.append(topAppsinGenres[i])
```

```python
plt.figure(figsize=(15,9))
plt.ylabel('Genres(App Count)')
plt.xlabel('Genres')
graph = sns.barplot(x=x,y=y,palette="deep")
graph.set_xticklabels(graph.get_xticklabels(), rotation=90, fontsize=12)
graph.set_title("Top Genres in the Playstore", fontsize = 20);
```

##### From the above visualization we can see that the Highest Number of Apps found in the Tools and Entertainment genres followed by Education, Medical and many more.
